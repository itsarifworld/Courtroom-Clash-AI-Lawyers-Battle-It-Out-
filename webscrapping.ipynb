```python
# LOOPING THROUGH ALL IPCs
law = []
description = []
for i in range(1, 100):
  url = f"https://devgan.in/ipc/section/{i}/"
  webpage = requests.get(url).text
  soup = BeautifulSoup(webpage, "lxml")

  # Check if h2 and p tags are found before accessing them
  h2_tags = soup.find_all('tr', class_="mys-head")
  p_tags = soup.find_all('p')

  if soup.find_all('tr', class_="mys-head"):
      law.append(soup.find_all('tr', class_="mys-head")[0].text)
  else:
      law.append(f"Section {i} - Title not found")

  if soup.find_all('tr', class_="mys-head"):
      description.append(soup.find_all('p')[0].text)
  else:
      description.append(f"Section {i} - Description not found")
```

```python
import pandas as pd

training_data = pd.DataFrame({
    "Title": law,
    "description": description,
    "case_type": ["general"] * len(law),
    "jurisdiction": ["India"] * len(law),
    "key_principles": [None] * len(law)      # we’ll enrich later
})

training_data.head()
```

```python
training_data['Title'] = training_data['Title'].str.replace(r"^S\.\s*\d+\s*", "", regex=True)
training_data.head()
```

```python
training_data.loc[training_data['Title'].str.contains("defamation", case=False), "case_type"] = "defamation"
training_data.loc[training_data['Title'].str.contains("murder", case=False), "case_type"] = "criminal"
training_data.loc[training_data['Title'].str.contains("property", case=False), "case_type"] = "property"
```

```python
training_data.to_csv('training_data.csv', index=False)
```
